```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format="retina"
plt.rcParams["text.usetex"]=True

import numpy as np
n, pi, delta=np.loadtxt("./../Desktop/M1source/learn_c/midterm1_MT.dat", unpack=True)
# n1, pi1, delta1=np.loadtxt("./../Desktop/M1source/learn_c/midterm1_random_exam.dat", unpack=True)

fig=plt.figure(figsize=(16, 8))
ax1= fig.add_subplot(121)
ax2= fig.add_subplot(122)
# ax3= fig.add_subplot(223)
# ax4= fig.add_subplot(224)


ax1.hlines(y=np.pi, xmin=0, xmax=1e+8, lw=2.5, color="k", label=r"$\pi$")
ax1.plot(n, pi, "o", markersize=5.0, color="r", label=r"$calculated \; \pi$")
ax2.plot(n, delta, "o", markersize=5.0, color="b", label=r"$\delta$")
ax2.plot(n, 3.16/np.sqrt(n), lw=2.5, color="k", label=r"$3.16/\sqrt{n}$")
# ax3.hlines(y=np.pi, xmin=0, xmax=1e+8, lw=2.5, color="k", label=r"$\pi$")
# ax3.plot(n1, pi1, "o", markersize=2.5, color="r", label=r"$calculated\; \pi \; by \; random$")
# ax4.plot(n1, delta1, "o", markersize=2.5, color="b", label=r"$\delta \; by \; random$")
# ax4.plot(n1, 2.54/np.sqrt(n1), lw=2.5, color="k", label=r"$2.54/\sqrt(n)$")

ax1.set_xscale("log")
ax2.set_xscale("log")
ax2.set_yscale("log")
# ax3.set_xscale("log")
# ax4.set_xscale("log")
# ax4.set_yscale("log")

ax1.set_xlabel(r"$n$", color="k", size=25)
ax1.set_ylabel(r"$\pi$", color="k", size=25)
ax2.set_xlabel(r"$n$", color="k", size=25)
ax2.set_ylabel(r"$\delta$", color="k", size=25)
# ax3.set_xlabel(r"$times$", color="k", size=25)
# ax3.set_ylabel(r"$\pi$", color="k", size=25)
# ax4.set_xlabel(r"$times$", color="k", size=25)
# ax4.set_ylabel(r"$\delta$", color="k", size=25)

ax1.legend(ncol=1, loc=4, borderaxespad=0, fontsize=25, frameon=True)
ax2.legend(ncol=1, loc=1, borderaxespad=0, fontsize=25, frameon=True)
# ax3.legend(ncol=1, loc=1, borderaxespad=0, fontsize=25, frameon=True)
# ax4.legend(ncol=1, loc=1, borderaxespad=0, fontsize=25, frameon=True)

ax1.set_title(r"$(2)$", fontsize=20)
ax2.set_title(r"$(3)$", fontsize=20)
ax1.tick_params(labelsize=20)
ax1.minorticks_on()
ax1.grid()
ax2.tick_params(labelsize=20)
ax2.minorticks_on()
ax2.grid()
plt.subplots_adjust(wspace=0.25)
#fig.tight_layout()

plt.savefig("./../Desktop/M1source/learn_c/midterm1graph.png")
plt.show()
```
