```python
#
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
%config InlineBackend.figure_format="retina"
plt.rcParams["text.usetex"]=True

import numpy as np
fig=plt.figure(figsize=(18, 8))

T=[1.00, 2.00, 3.00]
symbol=["-o", "D-", "x-"]
vibration=["damping", "critical", "overdamping"]

#0.0001
#x(t), v(t)2
for j in range(1, 3):
    ax1=fig.add_subplot("12{}".format(j))
    
    for i in range(0,3):
        time, x, v = np.loadtxt("./../Desktop/M1source/learn_c/dampingdata_{:.2f}_0.0001.dat".format(T[i]), unpack=True)
        plt.hlines(y = 0.0, xmin=0, xmax=90, linestyles="dashed", lw=2.5, color="k" )
        if(j == 1):
            plt.plot(time, x, "{}".format(symbol[i]), markersize=5, color=cm.jet(i/2), label=r"${}$".format(vibration[i]))
        else:
            plt.plot(time, v, "{}".format(symbol[i]), markersize=5, color=cm.jet(i/2), label=r"${}$".format(vibration[i]))
        
    
    plt.xlabel(r"$t/t_0(=\tilde{t})$", color="k", size=25)
    if(j == 1):
        plt.ylabel(r"$x(t)/a(=\tilde{x}(t))$", color="k", size=25)
    else:
        plt.ylabel(r"$v(t)t_0/a(=\tilde{v}(t))$", color="k", size=25)
    
    plt.xticks(color="k", size=25)
    plt.yticks(color="k", size=25)
    
    plt.legend(ncol=1, loc=1, borderaxespad=0, fontsize=20, frameon=False)
    
#fig.tight_layout()
plt.subplots_adjust(wspace=0.25)
plt.savefig("./../Desktop/M1source/learn_c/midterm2graph.png")
plt.show()
```
